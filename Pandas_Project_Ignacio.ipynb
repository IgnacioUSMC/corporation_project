```python
# prompt: then create a dataset where the following columns are created: purchse_date, using a random date in the past 10 months of 2024. store_id, a random number between 1 and 2500. order_id, a unique number where the store_id is concatenated with a random string. customer_id, based on a set of unique customer identifiers, per store_id. a list of products randomly from the list, 'inventory' with pricing and quantity (between 1 and 10), of a random length between 1 and 35. please include the previous code that includes the clothing

import pandas as pd
import random
from datetime import datetime, timedelta

# Previous code (clothing data generation)
size_adjectives = ["small", "medium", "large", "extra large", "petite", "plus size"]
clothing_categories = ["shirts", "pants", "dresses", "skirts", "jackets", "sweaters"]
clothing_brands = ["Nike", "Adidas", "Levi's", "Zara", "H&M", "Gucci", "Prada"]

products = []
for _ in range(100000):
    size = random.choice(size_adjectives)
    category = random.choice(clothing_categories)
    brand = random.choice(clothing_brands)
    products.append([size, category, brand])

df_products = pd.DataFrame(products, columns=['Size', 'Category', 'Brand'])


# Generate dataset
data = []
store_customers = {}  # Dictionary to store customer IDs per store

for _ in range(10000):
    purchase_date = datetime(2024, 1, 1) + timedelta(days=random.randint(0, 300))  # Random date in the past 10 months
    store_id = random.randint(1, 2500)
    order_id = str(store_id) + ''.join(random.choices('ABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789', k=8))

    if store_id not in store_customers:
        store_customers[store_id] = set()

    while True:
        customer_id = random.randint(1, 10000)
        if customer_id not in store_customers[store_id]:
            store_customers[store_id].add(customer_id)
            break

    num_products = random.randint(1, 35)
    inventory = []
    for _ in range(num_products):
        product_index = random.randint(0, len(df_products) - 1)
        quantity = random.randint(1, 10)
        price = round(random.uniform(10, 200), 2)  # Random price between $10 and $200

        inventory.append({'product': df_products.iloc[product_index].to_dict(), 'quantity': quantity, 'price': price})

    data.append({'purchase_date': purchase_date, 'store_id': store_id, 'order_id': order_id, 'customer_id': customer_id, 'inventory': inventory})


df_orders = pd.DataFrame(data)
df_orders
```

```python
df = pd.DataFrame (data, columns = ['purchase_date', 'store_id', 'order_id', 'customer_id', 'inventory'])
```

```python
# prompt: tell me which 10 stores have the most sales

# Calculate total sales per store
df_orders['total_sales'] = df_orders['inventory'].apply(lambda x: sum([item['quantity'] * item['price'] for item in x]))

# Group by store and sum total sales
store_sales = df_orders.groupby('store_id')['total_sales'].sum()

# Get the top 10 stores with the most sales
top_10_stores = store_sales.nlargest(10)

top_10_stores
```
